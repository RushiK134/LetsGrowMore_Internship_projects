In this task we have to convert a given image into its pencil sketch using Python. In this we use CV(Computer Version) library and Pillow module for the image processing.

**Importing Libraries**

```python
import numpy as np
import cv2
from PIL import Image
from IPython.display import display
```

Loding the Image

```python
img = cv2.imread("/content/my_img.jpg")
img = cv2.resize(img,(220,300))
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(rgb_img))
```

**Converting Image to Greyscale Image**

```python
greyscale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(greyscale_img))
```

**Converting the image into Negative Image**

```python
neg_img = cv2.bitwise_not(greyscale_img)
display(Image.fromarray(neg_img))
```

Smoothing the Image

```python
smooth_img = cv2.GaussianBlur(neg_img, (5,5),0,0)
display(Image.fromarray(smooth_img))
```

**Converting the image into Pencile Sketch**

```python
sketch_img = 255 - smooth_img
sketch = cv2.divide(greyscale_img, sketch_img, scale = 250.0)
display(Image.fromarray(sketch))
```

**Comparision Between Original image and Pencile Sketch**

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(rgb_img))
display(Image.fromarray(sketch))
```
