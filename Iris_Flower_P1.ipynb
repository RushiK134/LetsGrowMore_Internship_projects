Import Libraries

```python
#Import all the essencial libraries for out project

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

Loding the data set

```python
columns = ['Sepallength', 'Sepal width', 'Petal length', 'Petal width', 'Class_labels']
#Load the data
df = pd.read_csv('/content/IRIS.csv',names=columns)
df.head()
```

Visualization if our data Set

```python
df.describe()
```

```python
# Visualize the whole data
sns.pairplot(df,hue='Class_labels')
```

Seperating input columns and the output columns

```python
# seperate features and target.
data = df.values

X = data[:,0:4]
Y = data[:,4]
print(X)
```

Splitting the data into traning and testing

```python
# Splitting the data into traning and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
print(X_train)
```

MODEL1: Support vector machine algorithm

```python
from sklearn import model_selection
# Support vector machine algorithm
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X_train, y_train)
```

```python
prediction1 = model_svc.predict(X_test)
# Calculate the accutacy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction1)*100)
for i in range(len(prediction1)):
  print(y_test[i],prediction1[i])
```

MODEL2: LOGISTIC REGRESSION

```python
#  LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train, y_train)
```

```python
prediction2 = model_LR.predict(X_test)
# Calculate the accutacy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction2)*100)
for i in range(len(prediction2)):
  print(y_test[i],prediction2[i])
```

MODEL3: Decision Tree Classifier

```python
#  Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
model_DTC=DecisionTreeClassifier()
model_DTC.fit(X_train, y_train)
```

```python
prediction3 = model_DTC.predict(X_test)
# Calculate the accutacy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction3)*100)
for i in range(len(prediction3)):
  print(y_test[i],prediction3[i])
```

```python
# a detailed classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction2))
```
